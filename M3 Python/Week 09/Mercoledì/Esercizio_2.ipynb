```python
# Stampiamo un carattere per riga della stringa.
nome_scuola = "Epicode"
indice = 0

# Usiamo  il ciclo while.
# Usiamo len() per sapere quanto è lunga la stringa (in questo caso 7).
while indice < len(nome_scuola):
    print(nome_scuola[indice])
    indice += 1  # Spostiamoci al carattere successivo.

# Usiamo il ciclo for.
# Usiamo "carattere" come indice del ciclo.
for carattere in nome_scuola:
    print(carattere)
```

```python
# Calcoliamo le prime 10 potenze di 2, con il ciclo while

# Inizializziamo l'esponente a 0
esponente = 0
# Vogliamo le prime 10 potenze, quindi ci fermiamo quando arriviamo a 10 (escluso)
while esponente < 10:
    risultato = 2 ** esponente    # Calcoliamo la potenza e la stampiamo
    print(f"2^{esponente} = {risultato}")
    esponente += 1    # Incrementiamo l'esponente di 1 a ogni giro
```

```python
# 1. Chiediamo l'input all'utente
stringa = input("Inserisci una parola o una frase: ")

# 2. Calcoliamo la lunghezza per gestire i casi particolari
lunghezza = len(stringa)

if lunghezza >= 6:
    # Caso standard: estraiamo i primi 3 e gli ultimi 3 caratteri
    inizio = stringa[:3]    # Prende dall'indice 0 al 2
    fine = stringa[-3:]     # Prende gli ultimi 3 (da -3 alla fine)
    print(f"{inizio}...{fine}")

elif lunghezza > 0:
    # Caso ad hoc per stringhe corte (meno di 6 caratteri ma non vuote)
    # Possiamo decidere di mostrarla tutta o indicare che è troppo corta
    print(f"Stringa corta: {stringa}")

else:
    # Caso per stringa vuota
    print("Non hai inserito alcun carattere!")
```

```python
# Scomponiamo in fattori primi un numero ricevuto in input

numero_input = int(input("Inserisci un numero intero positivo: "))
n = numero_input # Creiamo una copia per non perdere il valore originale
fattori = []
divisore = 2

#  Metoto while
while n > 1:
    # Se il divisore divide n senza resto
    while n % divisore == 0:
        fattori.append(divisore) # Memorizziamo il fattore
        n //= divisore           # Dividiamo n e aggiorniamo il suo valore
    
    divisore += 1 # Passiamo al prossimo potenziale divisore

print(f"I fattori primi di {numero_input} sono: {fattori}")
```

```python
# Scomponiamo in fattori primi un numero ricevuto in input

numero_input = int(input("Inserisci un numero intero positivo: "))
n = numero_input
fattori = []

# Metodo for
# Testiamo tutti i possibili divisori da 2 fino al numero stesso
for divisore in range(2, numero_input + 1):
    # Finché il divisore attuale divide n, continua a memorizzarlo
    while n % divisore == 0:
        fattori.append(divisore)
        n //= divisore
    
    # Se n è diventato 1, abbiamo finito e possiamo uscire dal for
    if n == 1:
        break

print(f"Fattorizzazione completata: {fattori}")
```

```python
#Calcoliamo la potenza K di un numero N, entrambi definiti in input.

# Inizializziamo i parametri chiedendo all'utente
K = int(input("Inserisci la base K: "))
N = int(input("Quante potenze vuoi calcolare? (N): "))

potenze = []  # Lista vuota per memorizzare i risultati
esponente = 0 # Inizializzazione del contatore

# Usiamo il ciclo while
while esponente < N:
    valore = K ** esponente   # Calcolo
    potenze.append(valore)    # Memorizzazione in coda alla lista
    esponente += 1            # Incremento

# Stampa finale della lista
print(f"Lista delle prime {N} potenze di {K} (con while):")
print(potenze)
```

```python
#Calcoliamo la potenza K di un numero N, entrambi definiti in input.

K = int(input("Inserisci la base K: "))
N = int(input("Quante potenze vuoi calcolare? (N): "))

potenze = []

# Usiamo il ciclo for
for esponente in range(N):
    valore = K ** esponente
    potenze.append(valore)

# Stampa finale della lista
print(f"Lista delle prime {N} potenze di {K} (con for):")
print(potenze)
```

```python
# Data la lista "guadagni", calcore la media

guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]

# 1. Inizializziamo una variabile per accumulare la somma
somma_totale = 0

# 2. Ciclo for per attraversare ogni valore nella lista
for guadagno in guadagni:
    somma_totale += guadagno  # Aggiungiamo il valore attuale al totale

# 3. Calcoliamo la media dividendo per la lunghezza della lista
numero_mesi = len(guadagni)
media = somma_totale / numero_mesi

print(f"La somma totale dei guadagni è: {somma_totale}")
print(f"La media dei guadagni negli ultimi {numero_mesi} mesi è: {media}")
```

```python
# Data la lista "parole", asaminare tutti gli elementi e contare,
# per ogni elemento, quante sono le "e".

parole = [
    "Albergo", "Sedia", "Borgo", "Petalo", "Eremo", 
    "Belvedere", "Semestre", "Esteta", "Sosta", "Orpello", 
    "Abete", "Orologio", "Cesta", "Ermellino"
]

print("Conteggio della lettera 'e' (maiuscola e minuscola):")
print("-" * 30)

for parola in parole:
    # Trasformiamo la parola in minuscolo per contare tutte le 'e'
    conteggio = parola.lower().count('e')
    
    # Stampiamo il risultato usando una f-string
    print(f"Nella parola '{parola}' la lettera 'e' appare {conteggio} volte.")
```

```python
# Creiamo un dizionario che assegni ad ogni proprietario la sua auto
# Dopo controlliamo che auto guida Debbie

# 1. Creazione del dizionario
auto_proprietari = {
    "Ada": "Punto",
    "Ben": "Multipla",
    "Charlie": "Golf",
    "Debbie": "107"
}

# 2. Stampiamo il dizionario per intero
print("Dizionario completo:")
print(auto_proprietari)

# 3. Stampiamo l'auto associata a Debbie
# Usiamo la chiave "Debbie" tra parentesi quadre
print(f"\nL'auto guidata da Debbie è: {auto_proprietari['Debbie']}")
```

```python
# Creamo un dizionario di proprietari auto
# successivamente mostriamo le varie auto, tranne la Multipla

dizionario_auto = {
    "Ada": "Punto", 
    "Ben": "Multipla", 
    "Charlie": "Golf", 
    "Debbie": "107"
}

print("Auto disponibili (esclusa la Multipla):")

# Usiamo .values() per scorrere solo i nomi delle auto
for auto in dizionario_auto.values():
    # Controlliamo che l'auto NON sia una Multipla
    if auto != "Multipla":
        print(f"- {auto}")
```

```python
# Consideriamo 2 dizionari, aggiorniamo e analiziamo il risultato.

dizionario_auto = {
    "Ada": "Punto", 
    "Ben": "Multipla", 
    "Charlie": "Golf", 
    "Debbie": "107", 
    "Emily": "A1"
}

nuovi_proprietari = {
    "Ben": "Polo", 
    "Fred": "Octavia", 
    "Grace": "Yaris", 
    "Hugh": "Clio"
}

# Aggiorniamo il primo dizionario con il secondo
dizionario_auto.update(nuovi_proprietari)

# Stampiamo il risultato
print("Dizionario aggiornato:")
print(dizionario_auto)
```
