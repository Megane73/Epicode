```python
import pandas as pd

# Lista iniziale dei quantitativi degli oggetti venduti ogni mese
lst = [2000, 5500, 7200, 4320, 2500, 3900, 6410, 1280, 1900, 8150, 7100, 5350]

# Casting della lista in una Pandas Series
vendite = pd.Series(lst)

# Risposte per il CEO della Object SpA [cite: 43]
print(f"Vendita massima: {vendite.max()}, Vendita minima: {vendite.min()}")

# Filtro per le vendite maggiori di 5000 [cite: 45]
maggiori_5000 = vendite[vendite > 5000]
# Usiamo len() oppure il metodo .count() di Pandas
print(f"Vendite > 5000:\n{maggiori_5000.to_list()}\nIn totale sono {maggiori_5000.count()} mesi.")

# Filtro per le vendite minori di 3000 [cite: 46]
minori_3000 = vendite[vendite < 3000]
print(f"Vendite < 3000:\n{minori_3000.to_list()}")

# Media mensile [cite: 47]
print(f"Media mensile oggetti venduti: {vendite.mean().round(2)}")
```

```python
# La differenza principale tra un dizionario standard (fatturati_dict) e una Pandas Series (fatturati_series) risiede nelle loro funzionalità e nello scopo:
# Dizionario: È una struttura dati nativa di Python basata su coppie chiave-valore, non è ottimizzato per operazioni matematiche su larga scala.
# Series: È una struttura dati di Pandas, simile a un array monodimensionale, ma provvista di un indice (le chiavi/anni nel nostro caso).
# Cosa possiamo fare con la Series?
# Possiamo eseguire operazioni vettorializzate.
# Ad esempio, possiamo sommare o moltiplicare tutti i fatturati con un singolo comando (es: fatturati_series * 1.1 per applicare un aumento del 10%),
# calcolare medie o filtrare i dati direttamente tramite condizioni (es: fatturati_series[fatturati_series > 20000]),
# operazioni che richiederebbero dei cicli for se usassimo un dizionario.
```

```python
import pandas as pd

# Creazione della Series di Pandas (corrispondente all'array NumPy)
stipendi = pd.Series([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])

# Metodo 1: Ciclo for (Poco efficiente, sconsigliato con Pandas)
stipendi_raddoppiati_for = stipendi.copy()
for i in range(len(stipendi)):
    stipendi_raddoppiati_for[i] = stipendi[i] * 2

# Metodo 2: Masking (Il modo corretto e veloce in Pandas)
# Creiamo una maschera Booleana (True dove la condizione è verificata)
maschera = stipendi > 0

# Applichiamo la maschera per raddoppiare solo i valori corrispondenti a True
stipendi_raddoppiati_mask = stipendi.copy()
stipendi_raddoppiati_mask[maschera] = stipendi[maschera] * 2

# Metodo Bonus: Il modo più rapido (Broadcasting puro)
# In Pandas/NumPy basta fare: stipendi_veloce = stipendi * 2

print(stipendi_raddoppiati_mask.to_list())
```

```python
def tre_piu_grandi(lista_numeri):
    # Rimuoviamo i duplicati convertendo la lista in un set, poi la ordiniamo al contrario
    numeri_unici = sorted(list(set(lista_numeri)), reverse=True)
    
    if len(numeri_unici) < 3:
        return f"Lista troppo corta, ecco i maggiori disponibili: {numeri_unici}"
    else:
        return numeri_unici[:3]

print(tre_piu_grandi([10, 10, 8, 5, 12, 12, 1]))
print(tre_piu_grandi([5, 5]))
```

```python
def media_maggiori_k(lista, k):
    filtrati = [num for num in lista if num >= k]
    
    if len(filtrati) == 0:
        print(f"Nessun numero nella lista è maggiore o uguale a {k}.")
    else:
        media = sum(filtrati) / len(filtrati)
        print(f"La media dei numeri validi è: {media}")

media_maggiori_k([10, 20, 30, 40], 25)
```

```python
guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]

for i in range(len(guadagni)):
    mese_attuale = i + 1
    guadagno_attuale = guadagni[i]
    
    if i == 0:
        # Per il primo mese non c'è una media precedente
        print(f"Mese {mese_attuale}: {guadagno_attuale} €")
    else:
        # Calcoliamo la media di tutti gli elementi prima di 'i'
        media_prec = sum(guadagni[:i]) / i
        
        if guadagno_attuale > media_prec:
            confronto = "il guadagno attuale è maggiore"
        elif guadagno_attuale < media_prec:
            confronto = "il guadagno attuale è minore"
        else:
            confronto = "il guadagno attuale è uguale"
            
        print(f"Mese {mese_attuale}: {guadagno_attuale} € (media prec: {round(media_prec, 2)} € - {confronto})")
```
