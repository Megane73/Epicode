```python
import pandas as pd

df_diabetes = pd.read_csv("diabetes.csv")

# 1. Dimensioni e anteprima di 5 righe
print("Dimensioni:", df_diabetes.shape)
print(df_diabetes.head(5))

# 2. Metadati (tipi di dato, valori nulli)
df_diabetes.info()

# 3. Descrittori statistici (media, min, max, ecc.)
print(df_diabetes.describe())

# 4. Selezionare le fasce di età e calcolare la media della pressione
# Creiamo delle maschere booleane per filtrare
fascia_20_29 = df_diabetes[(df_diabetes['Age (years)'] >= 20) & (df_diabetes['Age (years)'] <= 29)]
fascia_30_39 = df_diabetes[(df_diabetes['Age (years)'] >= 30) & (df_diabetes['Age (years)'] <= 39)]
fascia_40_50 = df_diabetes[(df_diabetes['Age (years)'] >= 40) & (df_diabetes['Age (years)'] <= 50)]

# 5. Media della pressione sanguigna diastolica
print("Media pressione Diastolica 20-29:", fascia_20_29['Diastolic blood pressure (mm Hg)'].mean())
print("Media pressione Diastolica 30-39:", fascia_30_39['Diastolic blood pressure (mm Hg)'].mean())
print("Media pressione Diastolica 40-50:", fascia_40_50['Diastolic blood pressure (mm Hg)'].mean())
```

```python
import pandas as pd

df_insurance = pd.read_csv("insurance.csv")

# 1. Dimensioni, anteprima e colonne
print(df_insurance.shape)
print(df_insurance.head())
print(df_insurance.columns)

# 2. Medie di 'charges' rispetto a 'region'
print(df_insurance.groupby('region')['charges'].mean())

# 3. Medie rispetto a 'smoker' e 'sex'
print(df_insurance.groupby('smoker')['charges'].mean())
print(df_insurance.groupby('sex')['charges'].mean())

# 4. Descrittori statistici di 'bmi'
print(df_insurance['bmi'].describe())

# 5. Analisi charges rispetto ai quartili del bmi
# Sfruttiamo il raggruppamento dinamico di Pandas usando pd.qcut che divide in 4 fasce uguali (quartili)
df_insurance['bmi_quartile'] = pd.qcut(df_insurance['bmi'], q=4)
print(df_insurance.groupby('bmi')['charges'].agg(['min', 'mean', 'max']))
```

```python
import pandas as pd

df_pokemon = pd.read_csv("pokemon.csv")

# 1. Dimensioni e anteprima
print(df_pokemon.shape)
print(df_pokemon.head())
print(df_pokemon.columns)

# 2. CONFRONTO INDICI
# Stampiamo l'indice automatico di Pandas (0, 1, 2, 3...)
print("Indice automatico di Pandas:")
print(df_pokemon.index)

# Stampiamo la prima colonna del file (spesso è l'ID del Pokemon: 1, 2, 3...)
print("\nPrima colonna del dataset:")
print(df_pokemon.iloc[:, 0]) 

# RISPOSTA ALLA DOMANDA: "Combaciano?"
# Solitamente NO, perché l'indice di Pandas parte da 0, 
# mentre l'ID dei Pokemon parte da 1.

# SE NO, SETTARE LA PRIMA COLONNA COME INDICE:
# Usiamo il nome della colonna (che solitamente è '#' oppure 'id')
nome_prima_colonna = df_pokemon.columns[0]
df_pokemon.set_index(nome_prima_colonna, inplace=True)

print("\nNuovo indice impostato correttamente!")
print(df_pokemon.head())

# 3. Quali sono i leggendari?
leggendari = df_pokemon[df_pokemon['Legendary'] == True]
print(leggendari)

# 4. Leggendari di tipo Grass
leggendari_erba = leggendari[leggendari['Type 1'] == 'Grass']
print(leggendari_erba)

# 5. Leggendari di tipo Ice o Fire (usiamo l'operatore "or" che in Pandas è | )
leggendari_ghiaccio_fuoco = leggendari[(leggendari['Type 1'] == 'Ice') | (leggendari['Type 1'] == 'Fire')]
print(leggendari_ghiaccio_fuoco)

# 6. Trasformare la colonna Name nell'indice
df_pokemon = df_pokemon.set_index('Name')

# 7. Pokémon della prima generazione con Attack > 50 e HP < 60
gen_1_strani = df_pokemon[(df_pokemon['Generation'] == 1) & (df_pokemon['Attack'] > 50) & (df_pokemon['HP'] < 60)]
print(gen_1_strani)
```

```python
import pandas as pd

# 1. Caricamento con i parametri corretti
df_pub = pd.read_csv("Mappa-dei-pub-circoli-locali-in-Italia.csv", encoding="latin1", sep=";")

# 2. Totale dati e metadati
print("Totale dati:", df_pub.shape[0])
df_pub.info()

# 3. Primo, ultimo ed elemento casuale (usiamo .iloc che estrae per posizione)
print("Primo elemento:\n", df_pub.iloc[0])
print("\nUltimo elemento:\n", df_pub.iloc[-1])
print("\nElemento a caso:\n", df_pub.sample(1))

# 4. Anni di inserimento (ipotizziamo la colonna si chiami 'anno_inserimento')
print("\nAnni di inserimento presenti:", df_pub['Anno inserimento'].unique())

# 5. Contare le enoteche ed estrarne il nome. 
# Usiamo str.lower() e str.contains() per ovviare in modo semplice alla "sporcizia" del testo
enoteche = df_pub[df_pub['Nome'].str.lower().str.contains("enoteca", na=False)]
print(f"Ci sono {enoteche.shape[0]} enoteche.")
print("Ecco i nomi:\n", enoteche['Nome'].to_list())
```
